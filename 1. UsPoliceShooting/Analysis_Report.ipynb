```python
def get_stage(age : int) -> str:
    if age <= 12:
        return "child"
    elif age <= 19:
        return "Teen"
    elif age <= 40:
        return "Adult"
    elif age <= 60:
        return "Mid Aged"
    else:
        return "Senior citizen"
```

```python
df["life_stage"] = df.age.apply(lambda x: get_stage(x))
```
